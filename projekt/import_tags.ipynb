```python
import pandas as pd
tags = pd.read_csv(r"tags.csv")
df = pd.read_csv(r"dishes_tags.csv")
df_true_false = df.copy()
tags = tags[:30]
df = df[:80]

```

```python
display(df.head())
tags.head()
```

```python
for index, row in tags.iterrows():
    tag_id = row["id"]
    tag_name = row["name"]
    tag_type = row["type"]
    for index, row in df.iterrows():
        df[f"{tag_type}_{tag_name}"] = df["tags"].str.contains(str(tag_id)).astype(int)
        df_true_false[f"{tag_type}_{tag_name}"] = df["tags"].str.contains(str(tag_id))

```

```python
for index, row in tags.loc[tags['type'] == "cuisine"].iterrows():
    tag_id = row["id"]
    tag_name = row["name"]
    tag_type = row["type"]
    for index, row in df.iterrows():
        # if row["tags"].contains(str(tag_id)):
        if row["tags"] in str(tag_id):
            df.loc[f"{tag_type}", index] = tag_id
            
```
