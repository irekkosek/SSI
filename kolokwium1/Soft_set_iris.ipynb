```python

import numpy as np
import random as rn 
import pandas as pd
import seaborn as sns
```

```python
iris =pd.read_csv('iris.csv')
```

```python
class DataProcessing:
    @staticmethod
    def shuffle(x):
        for i in range(len(x)-1,0,-1):
            j=rn.randint(0,i-1)
            x.iloc[i], x.iloc[j] = x.iloc[j], x.iloc[i]
    @staticmethod
    def normalization(x):
        #x' = (x-min)/(max-min)
        values = x.select_dtypes(exclude=["object"]) #kopia zbioru x bez nazw kwiatÃ³w
        columnNames=values.columns.tolist() #nazwy kolumn z atrybutami
        for column in columnNames:
            data = x.loc[:,column] #wszystkie rekordy z jednej z kolumn
            max1=max(data)
            min1=min(data)
            for row in x.index: #range(0, len(x),1):
                xprim=(x.at[row,column]-min1)/(max1-min1)
                x.at[row,column]=xprim


    @staticmethod
    def split(x,k): #k=0.7 -> 70% do zbioru treningowego, 30% testowy
        split_index=int(k*len(x))
        df1 = x[:split_index] #divide dataframe by rows
        df2 = x[split_index:]
        
        return df1, df2
    @staticmethod
    def round_values(x):
        for column in x.select_dtypes(exclude=["object"]).columns:
            for row in x.index:
                x.at[row,column]=round(x.at[row,column],0)
    @staticmethod
    def calculate_positives(checked_vectors, validatingSet):
        positives = 0
        for tested_index in validatingSet.index:
                closest_variety = checked_vectors.at[tested_index, "variety"]
                true_variety = validatingSet.at[tested_index, "variety"]
                if(closest_variety == true_variety):
                    positives += 1
        return positives
    @staticmethod
    def calculate_accuracy(positives, validatingSet):
        acc=positives/len(validatingSet)
        return acc
                
        
```

```python
DataProcessing.shuffle(iris)
DataProcessing.normalization(iris)
DataProcessing.round_values(iris)
train, test = DataProcessing.split(iris,0.7)
# DataProcessing.round_values(train)
```

```python
class SoftSet:
    @staticmethod
    def calculate_soft_set_query(soft_set, query):
        result_df = soft_set.copy()
        results=[]
        for values in result_df.select_dtypes(exclude=["object"]).values:
            single_result=0
            for product_wage,request_wage in zip(values,query):
                single_result+= product_wage*request_wage
            results.append(single_result)
        result_df["fuzzy_result"]=results
        # display(result_df.sort_values(by=['fuzzy_result'], ascending=False))
        # display(result_df.loc[result_df['fuzzy_result'].idxmax(),"variety"])
        return result_df.loc[result_df['fuzzy_result'].idxmax(),"variety"] #max value in fuzzy_result column
        #this last line makes it less accurate (i was probably thinking about sort)
    @staticmethod
    def calculate_results(test, train):
        results_df=test.select_dtypes(exclude=["object"])
        results=[]
        for i in range(len(test)):
            results.append(SoftSet.calculate_soft_set_query(train, test.select_dtypes(exclude=["object"]).iloc[i])) #test
        results_df["variety"]=results
        display(results_df)
        return results_df
```

```python


results_df = SoftSet.calculate_results(train, test)

positives=DataProcessing.calculate_positives(results_df, test)
accuracy=DataProcessing.calculate_accuracy(positives, test)
print("Accuracy: ", accuracy)

```
